```python
import sys, os
import numpy as np
from sklearn.preprocessing import LabelBinarizer

default_font = { 'family' : 'Times New Roman',
              'weight' : 'normal',
              'size'   : 20,
    }

def flip90_left(arr):
    new_arr = np.transpose(arr)
    new_arr = new_arr[::-1]
    return new_arr

def flip90_right(arr):
    new_arr = arr.reshape(arr.size)
    new_arr = new_arr[::-1]
    new_arr = new_arr.reshape(arr.shape)
    new_arr = np.transpose(new_arr)[::-1]
    return new_arr

def flip180(arr):
    new_arr = arr.reshape(arr.size)
    new_arr = new_arr[::-1]
    new_arr = new_arr.reshape(arr.shape)
    return new_arr

# 计算RSSI_y和RSSI_x对应的距离比
def calc_distance_ratio(x, y, n=2):
    return 10**(-(y-x)/(10*n))

def calc_rssi_distance_ratio(vec,n=2, bPlt=False):
    m = len(vec)
    
    Ratio = np.zeros((m,m))
    for i in range(m):
        for j in range(m):
            if vec[i] > -99 and vec[j] > -99:
                Ratio[i,j] = calc_distance_ratio(vec[i], vec[j], n)
    if bPlt:
        X = range(m)
        Y = range(m)
        plt.figure(figsize=(10, 8))
        plt.contourf(X, Y, Ratio, cmap='jet')
        plt.colorbar()
    
    return Ratio

class FingerprintMap(object):
    '''
    指纹数据对象
    '''
    def __init__(self, mapdir):
        self.mapdir = mapdir
        self.bInited = False # 指纹地图是否初始化
        self.rawRssi = np.array([]) # 原始指纹数据
        self.rawCellID = np.array([])
        self.coordsList = np.array([])
        self.n = 0 # 指纹点个数
        self.n_ap = 0 # AP的个数
        self.type = 'RSSI' # 指纹类型，'RSSI', 'RATIO'
    
    def load_map(self):
        mapfile = ''+self.mapdir+'/train_sample-0.5s-all-2.txt'
        if not os.path.exists(mapfile):
            raise FileNotFoundError('No map file can be found!')
        rawData = np.loadtxt(mapfile, dtype='int')
        self.rawRssi = rawData[:,:-1]
        self.rawCellID = rawData[:,-1]
        
        self.n_ap = self.rawRssi.shape[1]
        
        print('number of AP =', self.n_ap)
        
        # 载入cell编号对应的坐标
        self.load_coords()
        
        self.n = self.coordsList.shape[0]
        
        print('number of fingerprint =', self.n)
        
        self.bInited = True
    
    def save_map(self):
        pass
    
    def load_coords(self):
        coordsfile = ''+self.mapdir+'/coords_mapping.txt'
        if not os.path.exists(coordsfile):
            raise FileNotFoundError('No coordinate file can be found!')
        self.coordsList = np.loadtxt(coordsfile, dtype='int')
        # print(self.coordsList)
        
    def get_normalized_rssi(self, normType):
        if not self.bInited:
            raise InitError('Not init the fingerprint!!')
        TINY_VAL = 1e-20 # 避免分子为0时的异常
        if normType == 'MinMax':
            min_rssi = self.rawRssi.min()
            max_rssi = self.rawRssi.max()
            norm_rssi = (self.rawRssi - min_rssi)/(max_rssi - min_rssi + TINY_VAL)
        elif normType == 'Normalize':
            mean_rssi = self.rawRssi.mean()
            std_rssi = self.rawRssi.std()
            norm_rssi = (self.rawRssi - mean_rssi) / (std_rssi + TINY_VAL)
        else:
            raise MethodTypeError('Not supported normalization methods!!')
            
        return norm_rssi
    
    def get_categorical_label(self):
        if not self.bInited:
            raise InitError('Not init the fingerprint!!')
        fp_id_dict = dict(zip(self.rawCellID, range(len(self.rawCellID))))
        lb = LabelBinarizer()
        lb.fit(list(fp_id_dict))
        tlabels = lb.transform(fp_id)
        
        return tlabels
    
    def generate_fingerprint_ratio(self, fp_id, bPlt=False):
        if not self.bInited:
            raise InitError('Not init the fingerprint!!')
            
        if fp_id < 1 or fp_id > self.n:
            raise RangeError('Error parameter of fp_id!!')
            
        # 获取所有fp_id的下标
        index = np.where(self.rawCellID == fp_id)
        if len(index) == 0:
            return
        
        rssi = self.rawRssi[index[0],:].astype('float')
        
        rssi[rssi == -99] = np.nan
        
        # 求取各个AP的平均值
        # print('rssi = ', rssi)
        mean_rssi = np.nanmean(rssi, axis=0)
        
        mean_rssi[np.isnan(mean_rssi)] = -99.0
        
        # print('mean_rssi = ',mean_rssi)
        
        # print(mean_rssi)
        
        # APx和APy对应的距离比
        Ratio = np.zeros((self.n_ap, self.n_ap))
        
        # 计算RSSI的距离比
        for m in range(self.n_ap):
            for n in range(self.n_ap):
                if mean_rssi[m] != -99 and mean_rssi[n] != -99:
                    Ratio[m, n] = calc_distance_ratio(mean_rssi[m], mean_rssi[n])
                
        if bPlt:
            X = range(self.n_ap)
            Y = range(self.n_ap)
            plt.figure(figsize=(10, 8))
            plt.contourf(X, Y, Ratio, cmap='jet')
            plt.colorbar()
            # plt.imshow(Ratio)
        return Ratio
    
    def generate_rssi_map(self, ap_id, bPlt=False):
        if not self.bInited:
            raise InitError('Not init the fingerprint!!')
            
        if ap_id < 1 or ap_id > self.n:
            raise RangeError('Error parameter of ap_id!!')
        # 获取AP为ap_id的RSSI数据    
        rssi = self.rawRssi[:,ap_id]
        # 获取对应的XY坐标值
        X = self.coordsList[:, 0]
        Y = self.coordsList[:, 1]
        # 根据坐标生成矩阵的下标
        min_X = X.min()
        min_Y = Y.min()
        
        print('min X and Y =', min_X, min_Y)
        iX = X - min_X
        iY = Y - min_Y
        
        
        print('iX, iY shape = ', iX.shape, iY.shape)
        
        # 生成map矩阵
        dimM = int(iX.max()+1)
        dimN = int(iY.max()+1)
        print('dimM, dimN =', dimM, dimN)
        rssi_map = np.zeros((dimM, dimN))-99
        rssi_map_std = np.zeros((dimM, dimN))
        
        # 循环处理每个指纹点
        cell_id = self.rawCellID
        for k in range(self.n):
            index = np.where(cell_id == k+1)
            if len(index) == 0:
                continue
            rssi_map[iX[k], iY[k]] = np.mean(rssi[index])
            rssi_map_std[iX[k], iY[k]] = np.std(rssi[index])
            
            
        if bPlt:
            #print('Get MAC: %s' %(self.coordsList))
            plt.figure(figsize=(10,8))
            x = range(dimM)
            y = range(dimN)
            X, Y = np.meshgrid(x, y)
            plt.contourf(X, Y, rssi_map.T,40, cmap="jet")
            plt.colorbar()
        
        return rssi_map, rssi_map_std
            
mapdir = mapdata_dir = '..//data//'
fpData = FingerprintMap(mapdir)
fpData.load_map()

font = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 26,
        }
label_font = {'family' : 'Times New Roman',
        #'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 26,
        }

AP_ID = 11
rssi_map, rssi_map_std = fpData.generate_rssi_map(AP_ID, bPlt=True)

AP_ID = 21
rssi_map, rssi_map_std = fpData.generate_rssi_map(AP_ID, bPlt=True)
#print('Get MAC: %s' %(self.coordsList))
fig,ax=plt.subplots(figsize=(10,8))
x = range(rssi_map.shape[0])
y = range(rssi_map.shape[1])
X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, flip90_right(rssi_map),40, cmap="jet")
cb=plt.colorbar()
cb.set_label('mean RSSI',fontdict=font) #设置colorbar的标签字体及其大小

plt.xlabel('X/m',label_font)
plt.ylabel('Y/m',label_font)
plt.tick_params(labelsize=23)
labels = ax.get_xticklabels() + ax.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]
        
fig = plt.gcf()
fig.savefig('../figure/AP_19_rssi_distrib.png', dpi=600)

#AP_ID = 10
#rssi_map, rssi_map_std = fpData.generate_rssi_map(AP_ID, bPlt=True)

FP_ID = 40
fp_rssi_ratio = fpData.generate_fingerprint_ratio(FP_ID, bPlt=True)
plt.figure(figsize=(10,8))
x = range(25)
y = range(25)
X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, fp_rssi_ratio[0:25,0:25].T,40, cmap="jet")
plt.xlabel('AP index',label_font)
plt.ylabel('AP index',label_font)
cb=plt.colorbar()

cb.set_label('RSSI ratio',fontdict=font) #设置colorbar的标签字体及其大小
fig = plt.gcf()
fig.savefig('../figure/FP_40_ratio_map.png', dpi=600)
```

### 蓝牙信号质量评估
蓝牙信号在空间的传播可采用如下模型表示：
$$r=-(10nlog(d)+a$$
其中，$r$是采集的RSSI（单位dBm），$d$为蓝牙节点到接收端的距离（单位为m），$a$为环境相关的因子。

定义任意两个蓝牙信号的RSSI距离比为：
$$d(p, q) = -10^{-\frac{r_q - r_p}{10n}} $$

则在某一个点上接收到的所有蓝牙节点(共M个）的RSSI之间可形成一个距离比矩阵：
$$ D(L) = [d(i,j)]_{M \times M} $$
其中，$i,j=1,\cdots,M$

```python
from math import sin, cos, radians, sqrt
from filterpy.kalman import KalmanFilter
from filterpy.common import Q_discrete_white_noise

def initialize_filter(kf, std_R=None):
    """ helper function - we will be reinitialing the filter
    many times.
    """
    # kf.x.fill(0.)
    kf.P = np.eye(kf.dim_x) * .1
    if std_R is not None:
        kf.R = np.eye(kf.dim_z) * std_R

def make_cv_filter(dt, std):
    cvfilter = KalmanFilter(dim_x = 2, dim_z=1)
    cvfilter.x = np.array([0., 0.])
    cvfilter.P *= 100
    cvfilter.R *= std**2
    cvfilter.F = np.array([[1, dt],
                           [0,  1]], dtype=float)
    cvfilter.H = np.array([[1, 0]], dtype=float)
    cvfilter.Q = Q_discrete_white_noise(dim=2, dt=dt, var=0.02)
    return cvfilter
def run_cv_filter(zs_2, std, Rc, x0):
    cvfilter = make_cv_filter(dt, std=std)
    cvfilter.x = x0
    z_xs2 = zs_2

    # reinitialize filter
    initialize_filter(cvfilter)
    cvfilter.R = np.eye(1)*Rc
    
    xs = []
    for z in z_xs2:
        cvfilter.predict()
        
        # print('residuals = ', np.abs(z-cvfilter.x[0]))
        if np.abs(z-cvfilter.x[0]) < 15:
            cvfilter.update([z])            
        xs.append(cvfilter.x[0])
    return np.array(xs)

pos = est_pos2
dt = 0.5
xk = run_cv_filter(pos[:, 0], 2, 5, np.array([pos[0,0],0]));
yk = run_cv_filter(pos[:, 1], 2, 5, np.array([pos[0,1],0]));
```

```python
import datetime, time

class IMU(object):
    def __init__(self):
        pass

def time2sec(y):
    '''
    时间类型时分秒转换成秒
    '''
    h = y.hour  #直接用datetime.time模块内置的方法，得到时、分、秒
    m = y.minute
    s = y.second
    return int(h)*3600 + int(m)*60 + int(s) #int()函数转换成整数运算

def ms_time_parser(s):
    return float(s[-3:])/1000

def time_parser(s):
    return datetime.datetime.strptime(s[5:-4],'%Y-%m-%d %H:%M:%S')

def xyz_parser(s):
    ss = s.split(':')
    
    return float(ss[1])

def load_imu_data(filename):
    if not os.path.exists(filename):
        raise FileNotFoundError('No acc file found!!')
        
    data = pd.read_csv(filename, delimiter=',|;', names=['Time', 'X', 'Y', 'Z'])
    
    data['Ms'] = data['Time'].apply(ms_time_parser)
    data['Time'] = data['Time'].apply(time_parser)
    data['Ms'] = data['Time'].apply(time2sec) + data['Ms']
    data['X'] = data['X'].apply(xyz_parser)
    data['Y'] = data['Y'].apply(xyz_parser)
    data['Z'] = data['Z'].apply(xyz_parser)

    return data

acc_file = '../data/20191104091342accelerometer.txt'
acc = load_imu_data(acc_file)

gyo_file = '../data/20191104091342gyroscope.txt'
gyo = load_imu_data(gyo_file)

mag_file = '../data/20191104091342magnetometer.txt'
mag = load_imu_data(mag_file)

acc.head()
```

### 通过插值进行数据同步
实际使用时，应进行时间同步。插值处理的合理性有待探讨，暂时先这么处理。


```python
from scipy.interpolate import interp1d

def sync_imu_mag(acc, gyo, mag, t_sample=1/200):
    '''
    acc,gyo,mag:
        DataFrame类型的数据，包含Time、X、Y、Z,Ms共5列数据
    '''
    t_acc = acc['Ms']
    t_gyo = gyo['Ms']
    t_mag = mag['Ms']
    
    # t_sample = 1/200
    
    t_min = np.max([np.min(t_acc[:]), np.min(t_gyo[:]), np.min(t_mag[:])])
    t_max = np.min([np.max(t_acc[:]), np.max(t_gyo[:]), np.max(t_mag[:])])
    t_interp = np.linspace(t_min+t_sample, stop=t_max-t_sample, num=int((t_max-t_min)/t_sample))

    # store the interpolation 
    sync_data = pd.DataFrame()
    
    #Time
    sync_data['Ms'] = t_interp
    
    # ACC
    acc_x = acc['X']
    acc_y = acc['Y']
    acc_z = acc['Z']
    f = interp1d(t_acc, acc_x,kind="cubic")

    sync_data['ACCx'] = f(t_interp)
    f = interp1d(t_acc, acc_y,kind="cubic")
    sync_data['ACCy'] = f(t_interp)
    f = interp1d(t_acc, acc_z,kind="cubic")
    sync_data['ACCz'] = f(t_interp)
    
    # GYO
    gyo_x = gyo['X']
    gyo_y = gyo['Y']
    gyo_z = gyo['Z']
    f = interp1d(t_gyo, gyo_x,kind="cubic")
    sync_data['GYOx'] = f(t_interp)
    f = interp1d(t_gyo, gyo_y,kind="cubic")
    sync_data['GYOy'] = f(t_interp)
    f = interp1d(t_gyo, gyo_z,kind="cubic")
    sync_data['GYOz'] = f(t_interp)
    
    # MAG
    mag_x = mag['X']
    mag_y = mag['Y']
    mag_z = mag['Z']
    print(mag_x.shape, t_mag.shape)
    f = interp1d(t_mag, mag_x,kind="cubic")
    sync_data['MAGx'] = f(t_interp)
    f = interp1d(t_mag, mag_y,kind="cubic")
    sync_data['MAGy'] = f(t_interp)
    f = interp1d(t_mag, mag_z,kind="cubic")
    sync_data['MAGz'] = f(t_interp)
    
    return sync_data
```

### 采用madgwick滤波算法进行方位角估计
滤波更新中，加速度和磁力计的单位可任意（处理时进行了归一化，从而单位无关），陀螺仪需要转换为弧度/s

```python
### 使用madgwick滤波计算方位角
from ahrspy.madgwickahrs import MadgwickAHRS
import math


ahrs = MadgwickAHRS(sampleperiod=dt, beta=0.1)

# update using the measurement
deg2rad = math.pi/180.0

gyo_new = sync_imu_data.loc[:, ['GYOx','GYOy','GYOz']].values * deg2rad
acc_new = sync_imu_data.loc[:, ['ACCx', 'ACCy', 'ACCz']].values
mag_new = sync_imu_data.loc[:, ['MAGx', 'MAGy', 'MAGz']].values

n_sample = sync_imu_data.shape[0]
att = np.zeros((n_sample,3))
for k in range(n_sample):
    gyroscope = gyo_new[k, :]
    accelerometer = acc_new[k, :]
    magnetometer = mag_new[k, :]
    # ahrs.update(gyroscope, accelerometer, magnetometer)
    ahrs.update_imu(gyroscope, accelerometer)
    roll, pitch, yaw = ahrs.quaternion.to_euler_angles()
    att[k, 0] = roll
    att[k, 1] = pitch
    att[k, 2] = yaw
```
