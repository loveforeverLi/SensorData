```python
import sys, os
import numpy as np
import datetime, time
from sklearn.preprocessing import LabelBinarizer

def time2sec(y):
    '''
    时间类型时分秒转换成秒
    '''
    h = y.hour  #直接用datetime.time模块内置的方法，得到时、分、秒
    m = y.minute
    s = y.second
    return int(h)*3600 + int(m)*60 + int(s) #int()函数转换成整数运算

def time_parser(s):
    return datetime.datetime.strptime(s[0:-4],'%Y-%m-%d %H:%M:%S')

def ms_time_parser(s):
    return float(s[-3:])/1000

def pd_parse_time(pd_data):
    pd_data['TSEC'] = pd_data['TIME'].apply(ms_time_parser)
    pd_data['TIME'] = pd_data['TIME'].apply(time_parser)
    pd_data['TSEC'] = pd_data['TSEC'] + pd_data['TIME'].apply(time2sec)


f2_dis_rssi_file = '../data/fn2_ap_selected_rssi_distance.csv'

drssi = pd.read_csv(f2_dis_rssi_file, header=None, names=['TIME','FID','FX','FY','AID','MAC','AX','AY','AZ','RSSI','DIST'])

pd_parse_time(drssi)

```

```python
from scipy.optimize import curve_fit

def fitfunc(logd, a, n):
    return  -10*n*logd+a

max_aid = drssi['AID'].max()
min_aid = drssi['AID'].min()

print('min/max mac id', min_aid, max_aid)

font = {'family' : 'serif',
        'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 16,
        }
label_font = {'family' : 'Times New Roman',
        #'color'  : 'darkred',
        'weight' : 'normal',
        'size'   : 26,
        }

# 获得所有蓝牙节点信号传播规律
ApInfo = pd.DataFrame(columns=('MAC','a','n','x','y','z'))
for k in range(min_aid, max_aid+1):
    k_index = drssi.index[drssi['AID'] == k].tolist()
    if not k_index:
        continue
    
    k_rssi = drssi.loc[k_index, ['RSSI']]
    k_dist = drssi.loc[k_index, ['DIST']]
    
    # print('data length =', len(k_rssi), 'MAC =', drssi.loc[k_index[0], ['MAC']])
    
    # 数据拟合
    x = np.log10((k_dist[:].values.T)[0])
    y = (k_rssi[:].values.T)[0]
    
    ind = np.argsort(x)
    x = x[ind]
    y = y[ind]
    
    # fit
    popt, pcov = curve_fit(fitfunc, x, y)
    
    # coef
    a = popt[0]
    n = popt[1]
    
    fval = fitfunc(x, a, n)
    
    # print('a = {%f}, n = {%f}' %(a, n))
    
    ApInfo = ApInfo.append(pd.DataFrame({'MAC' : drssi.loc[k_index[0], ['MAC']].values, \
    'a' : a, \
    'n' : n, \
    'x' : drssi.loc[k_index[0], ['AX']].values, \
    'y' : drssi.loc[k_index[0], ['AY']].values, \
    'z' : drssi.loc[k_index[0], ['AZ']].values}),ignore_index=True)
    # 画图：
    if k==10:    
        fig, ax = plt.subplots(figsize=(10,8))
        plt.scatter(10**x, y, c='r',label='Measured, AP ID'+str(k))
        plt.plot(10**x, fval, 'bo-', label='Fitted curve')
        plt.tick_params(labelsize=23)
        labels = ax.get_xticklabels() + ax.get_yticklabels()
        [label.set_fontname('Times New Roman') for label in labels]
        # ax.set(xlabel='Real distance(m)', ylabel='RSSI (dBm)', title='RSSI vs Distance for AP device 1')
        plt.xlabel('Real distance/m', label_font)
        plt.ylabel('RSSI/dBm', label_font)
        plt.title('RSSI vs Distance for AP device 10',label_font)
        
        plt.legend(prop=label_font)
        
        fig = plt.gcf()
        fig.savefig('../figure/RSSI_fitting.png', dpi=600)
# 固定位置处的信号
n = 1
for j in range(1,n):
    j_index = drssi.index[drssi['FID'] == j].tolist()
    j_index2 = drssi.index[(drssi['FID'] == j) & (drssi['AID'] == 10)].tolist()
    
    j_time = drssi.loc[j_index2, ['TSEC']]
    j_rssi = drssi.loc[j_index2, ['RSSI']]
    j_dist = drssi.loc[j_index2, ['DIST']]
    j_Mac = drssi.loc[j_index2, ['MAC']]
    
    fig, ax = plt.subplots(figsize=(10,8))
    plt.plot(j_time, j_rssi, 'bo-', label='FID='+str(j))
    ax.set(xlabel='Real distance(m)', ylabel='RSSI (dBm)', title='RSSI vs Distance for Fingerprint Point')
    #plt.xlim(50550,50552)
    plt.legend(prob=label_font)
    
    plt.figure(figsize=(10,8))
    plt.hist(x=j_rssi.T, bins=int(j_rssi.max()-j_rssi.min()), density=False)
    plt.show()
```

### 构建基于距离的指纹
根据几何距离构建指纹，指纹是唯一的，不存在漂移。

构建方法：
对每个指纹点，计算所有MAC节点到该指纹点的距离

```python
radioMap = pd.DataFrame()
min_fid = drssi['FID'].min()
max_fid = drssi['FID'].max()

nfid = max_fid - min_fid + 1
naid = max_aid - min_aid + 1
radioData = np.zeros((nfid, naid+1))
for k in range(nfid):
    radioData[k, 0] = k+min_fid
    for j in range(naid):
        index = drssi.index[(drssi['FID']==k+min_fid) & (drssi['AID']==j+min_aid)]
        
        if index.any():
            # print(drssi.loc[index[0], ['DIST']].values)
            radioData[k, j+1] = drssi.loc[index[0], ['DIST']].values

radioMap['FID'] = radioData[:, 0]
for j in range(naid):
    radioMap['AID'+str(j+1)] = radioData[:,j+1]
```

### 测试数据处理
通过获得a和n得到距离值。
每秒内的RSSI数据进行平均（or滤波？），然后求出距离值，再根据WKNN算法求解

```python
test_file = '../data/fn2_ap_selected_rssi_distance_test.csv'

test_data = pd.read_csv(test_file, header=None, names=['TIME','FID','FX','FY','AID','MAC','AX','AY','AZ','RSSI','DIST'])

pd_parse_time(test_data)

test_data.head()
```

```python
t = np.floor(test_data['TSEC'].values)

t_min = np.min(t)
t_max = np.max(t)

def match_radioMap(x, M):
    nfid = int(M.shape[0])
    c = np.zeros((1, nfid))+np.nan
    
    ind = np.where(x==0)
    # print('ind = ', ind)
    x[np.where(x==0)]=np.nan
    for k in range(nfid):
        # print(k)
        m = np.array(M.loc[k, :].values)
        # print(x-m)
        dx = x-m
        # print('dx=',dx)
        dx[ind] = 0
        # print('dx=', dx)
        c[0, k] = np.linalg.norm(dx)
        # print(k, c[0, k])
        
    return c

est_pos = np.zeros((int(t_max-t_min)+1, 2))
for k in range(int(t_min), int(t_max)+1):
    ind = (np.where(t == k))[0]
    if len(ind) < 4:
        continue
        
    calc_dist_data = np.zeros((1, 26))
    calc_dist_count = np.zeros((1,26))
    for j in range(len(ind)):
        aid = test_data.loc[ind[j], 'AID']
        rssi = test_data.loc[ind[j], 'RSSI']
        a = ApInfo.loc[aid-1, 'a']
        n = ApInfo.loc[aid-1, 'n']
        # print('j=',j,aid, a, n, rssi, 10**((a-rssi)/(10*n)))
        calc_dist_data[0, aid-1] = calc_dist_data[0, aid-1] + 10**((a-rssi)/(10*n))
        calc_dist_count[0, aid-1] = calc_dist_count[0, aid-1] + 1
    
    for j in range(26):
        if calc_dist_count[0, j] > 1:
            # print('j =',j, calc_dist_data[0, j], calc_dist_count[0, j])
            calc_dist_data[0, j] /= calc_dist_count[0, j]
    
    match = match_radioMap(calc_dist_data[0], radioMap)
    
    # print(match[0])
    
    p_index = np.argpartition(match[0], 1)[:2]
    
    
    if k==int(t_min)+100:
        plt.plot(match.T, 'o-', label='Match')
        plt.plot(p_index, match[0, p_index], '*', label='Largest')
        plt.legend()
        # plt.xlim(0, 100)
        plt.show()
    
    
    # print(k, p_index)
    est_pos[k-int(t_min),0] = np.dot(match[0, p_index]/np.sum(match[0, p_index]), coordsList[p_index,0]);
    est_pos[k-int(t_min),1] = np.dot(match[0, p_index]/np.sum(match[0, p_index]), coordsList[p_index,1]);
    
# print('est pos = ', est_pos)
```

### KF filter


```python
from math import sin, cos, radians, sqrt
from filterpy.kalman import KalmanFilter
from filterpy.common import Q_discrete_white_noise

def initialize_filter(kf, std_R=None):
    """ helper function - we will be reinitialing the filter
    many times.
    """
    # kf.x.fill(0.)
    kf.P = np.eye(kf.dim_x) * .1
    if std_R is not None:
        kf.R = np.eye(kf.dim_z) * std_R

def make_cv_filter(dt, std):
    cvfilter = KalmanFilter(dim_x = 2, dim_z=1)
    cvfilter.x = np.array([0., 0.])
    cvfilter.P *= 100
    cvfilter.R *= std**2
    cvfilter.F = np.array([[1, dt],
                           [0,  1]], dtype=float)
    cvfilter.H = np.array([[1, 0]], dtype=float)
    cvfilter.Q = Q_discrete_white_noise(dim=2, dt=dt, var=0.02)
    return cvfilter
def run_cv_filter(zs_2, std, Rc, x0):
    cvfilter = make_cv_filter(dt, std=std)
    cvfilter.x = x0
    z_xs2 = zs_2

    # reinitialize filter
    initialize_filter(cvfilter)
    cvfilter.R = np.eye(1)*Rc
    
    xs = []
    for z in z_xs2:
        cvfilter.predict()
        if abs(z-cvfilter.x[0]) < 15:
            cvfilter.update([z])
        xs.append(cvfilter.x[0])
        
    return np.array(xs)

pos = est_pos
dt = 1
xk = run_cv_filter(pos[:, 0], 2, 5, np.array([pos[0,0],0]));
yk = run_cv_filter(pos[:, 1], 2, 5, np.array([pos[0,1],0]));
```

```python
# 计算RSSI_y和RSSI_x对应的距离比
def calc_distance_ratio(x, y, n=2):
    return 10**(-(y-x)/(10*n))

def calc_rssi_distance_ratio(vec,n=2, bPlt=False):
    m = len(vec)
    
    Ratio = np.zeros((m,m))
    for i in range(m):
        for j in range(m):
            if vec[i] > -99 and vec[j] > -99:
                Ratio[i,j] = calc_distance_ratio(vec[i], vec[j], n)
    if bPlt:
        X = range(m)
        Y = range(m)
        plt.figure(figsize=(10, 8))
        plt.contourf(X, Y, Ratio, 40, cmap='jet')
        plt.colorbar()
    
    return Ratio
```
